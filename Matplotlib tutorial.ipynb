```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Graph in 2D")
plt.savefig('firstplot.png')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sine Waveform')
plt.plot(x,y)
```

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y,align='center')
plt.bar(x2,y2, color='g')

plt.title('Bar graph')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,20,51,5,79,31,27])
plt.hist(a,bins=10)
plt.title('Histogram')
```

```python
data = [np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```
